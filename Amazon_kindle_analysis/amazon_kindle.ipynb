```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('kindle.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df=df.drop(['imgUrl','productURL'],axis=1)
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['author'].fillna('Unknown',inplace=True)
df['soldBy'].fillna('Unknown',inplace=True)
df['publishedDate'].fillna('Unknown',inplace=True)
```

```python
df.isnull().sum()
```

 Authors with Most Books On Amazon Kindle


```python
books_by_author=df[df['author'] != 'Unknown']['author'].value_counts().head(10)
fig,ax=plt.subplots(figsize=(12,6))
ax.bar(books_by_author.index,books_by_author.values)
plt.xticks(rotation=90)
plt.show()
ax.set_title('Top 10 Authors')
ax.set_xlabel('Authors')
ax.set_ylabel('Number of Books')
plt.show()
```

Top Ten Genres

```python
top_genres=df['category_name'].value_counts().head(10)
fig,ax=plt.subplots(figsize=(12,6))
ax.bar(top_genres.index,top_genres.values,color='green')
plt.xticks(rotation=90)
ax.set_title('Top 10 Genres')
ax.set_xlabel('Genres')
ax.set_ylabel('Number of Books')
```

Distribution of Price

```python
fig,ax=plt.subplots(figsize=(12,6))
ax.hist(df['price'],bins=20)
ax.set_xlim(0,300)
ax.set_title('Price Distribution')
ax.set_xlabel('Price')
ax.set_ylabel('Number of Books')
```

Distribution of Stars

```python
fig,ax=plt.subplots(figsize=(12,6))
ax.hist(df['stars'],bins=20)
ax.set_label('Distribution of Stars')
ax.set_xlabel('Stars')
ax.set_ylabel('Number of Books')
```

Category wise best seller

```python
popular_genres=df[df['isBestSeller'] == True]['category_name'].value_counts().head(10)
fig,ax=plt.subplots(figsize=(12,6))
ax.bar(popular_genres.index,popular_genres.values)
plt.xticks(rotation=90)
ax.set_title('Top 10 Popular Genres of Best Sellers')
ax.set_xlabel('Genres')
ax.set_ylabel('Number of Books')
```

```python
avg_price=df.groupby('category_name')['price'].mean().sort_values(ascending=False).head(10)
fig,ax=plt.subplots(figsize=(12,6))
ax.bar(avg_price.index,avg_price.values)
plt.xticks(rotation=90)
ax.set_title('Average Price of Books by Category')
ax.set_xlabel('Categories')
ax.set_ylabel('Average Price')
```

```python
author_rating=df.groupby('author')['stars'].mean().sort_values(ascending=False).head(10)
fig,ax=plt.subplots(figsize=(12,6))
ax.barh(author_rating.index,author_rating.values)
plt.xticks(rotation=90)
ax.set_title('Top 10 Authors by Average Rating')
ax.set_xlabel('Average Rating')
ax.set_ylabel('Authors')
plt.show()
```

```python
top10_reviews = df.sort_values('reviews', ascending=False).head(10)[['title', 'reviews']].set_index('title')
fig, ax = plt.subplots(figsize=(10, 6))
top10_reviews.plot(kind='bar', ax=ax, color='skyblue')
ax.set_title("Top 10 Most Reviewed Books")
ax.set_xlabel("Book Title")
ax.set_ylabel("Number of Reviews")
ax.tick_params(axis='x', rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x='price', y='stars', data=df)
ax.set_title("Price vs Stars")
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(x='isKindleUnlimited', y='price', data=df)
ax.set_title("Kindle Unlimited vs Price")
```

```python
sns.countplot(data=df, x='isEditorsPick', hue='isBestSeller')
plt.title("Does Editor's Picks Become Best Sellers?")
plt.show()
```
